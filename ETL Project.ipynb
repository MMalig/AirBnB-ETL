```python
listings_detail_file = listings_detail_file[[
'id',
'listing_url',
'name',
'summary',
'space',
'description',
'host_id',
'host_name',
'host_location',
'host_response_time',
'host_listings_count',
'neighbourhood_cleansed',
'zipcode',
'smart_location',
'property_type',
'room_type',
'accommodates',
'bathrooms',
'bedrooms',
'beds',
'bed_type',
'amenities',
'price',
'number_of_reviews',
'review_scores_rating']]
```

```python
listings_detail_file['amenities']=listings_detail_file['amenities'].str.replace("{","", regex=True)
listings_detail_file['amenities']=listings_detail_file['amenities'].str.replace('"','', regex=True)
listings_detail_file['amenities']=listings_detail_file['amenities'].str.replace(',',', ', regex=True)
```

```python
listings_detail_file['price']=listings_detail_file['price'].str.replace('$','', regex=True)
listings_detail_file['price']=listings_detail_file['price'].str.replace(',','', regex=True)
```

```python
listings_detail_file['name']=listings_detail_file['name'].fillna(value='N/A')
listings_detail_file['summary']=listings_detail_file['summary'].fillna(value='N/A')
listings_detail_file['space']=listings_detail_file['space'].fillna(value='N/A')
listings_detail_file['description']=listings_detail_file['description'].fillna(value='N/A')
listings_detail_file['host_name']=listings_detail_file['host_name'].fillna(value='N/A')
listings_detail_file['host_location']=listings_detail_file['host_location'].fillna(value='N/A')
listings_detail_file['host_response_time']=listings_detail_file['host_response_time'].fillna(value='N/A')
listings_detail_file['host_listings_count']=listings_detail_file['host_listings_count'].fillna(value=0.0)
listings_detail_file['bathrooms']=listings_detail_file['bathrooms'].fillna(value=0.0)
listings_detail_file['zipcode']=listings_detail_file['zipcode'].fillna(value='N/A')
listings_detail_file['bathrooms']=listings_detail_file['bathrooms'].fillna(value=0.0)
listings_detail_file['bedrooms']=listings_detail_file['bedrooms'].fillna(value=0.0)
listings_detail_file['beds']=listings_detail_file['beds'].fillna(value=0.0)
listings_detail_file['review_scores_rating']=listings_detail_file['review_scores_rating'].fillna(value=0)
```
